```python
def tuple2hex(c):
    c = tuple(map(lambda x: round(256*x), c))
    return '#%02x%02x%02x' % c
```

## Load character names

```python
# Read characters
relationship_df = pd.read_csv("dataframes/relationships_1 gardens_of_the_moon.csv")
print(relationship_df.value.sum())
print(relationship_df.size)
# take only characters from the current book
#relationship_df = relationship_df.loc[relationship_df["book"] == "Gardens of the Moon"]

# remove sparse connections (under a certain threshold)
min_occurances = 0.000*relationship_df.value.sum()
relationship_df = relationship_df.loc[relationship_df["value"]>=min_occurances]
```

```python
relationship_df.size
```

## Graph analysis and visualization

```python
# Create a graph from a pandas dataframe
relationship_size = 1000000
relationship_size = min(relationship_size,relationship_df.size)
G = nx.from_pandas_edgelist(relationship_df[0:min(relationship_size,relationship_df.size)], 
                            source = "source", 
                            target = "target", 
                            edge_attr = "value", 
                            create_using = nx.Graph())
relationship_df.size
```

```python
a = nx.adjacency_matrix(G, weight=G.edge_attr_dict_factory)
#print(a[4])
#plt.imshow(a, cmap='hot', interpolation='nearest')
#print(G.edges.values)
```

```python
nodelist = list(G.nodes)
A = nx.to_numpy_array(G, nodelist=nodelist)

#mask = np.zeros_like(A)
#mask[np.tril_indices_from(mask)] = True
#sns.heatmap(A, mask=mask, linewidth=0.5)

sns.heatmap(A)
heatmap(A, cbar=False)
```

```python
plt.figure(figsize=(10,10))
pos = nx.spiral_layout(G)
nx.draw(G, with_labels=True, node_color='skyblue', edge_cmap=plt.cm.Blues, pos = pos)
plt.show()
```
